```python
def move(x,y,a,b,n):
    '''
    A function that lists all possible movements for a knight
    that starts from (x,y) point with (a,b) movements in an
    nxn grid
    '''
    # All movements
    m1 = (x+a, y+b)
    m2 = (x-a, y+b)
    m3 = (x+a, y-b)
    m4 = (x-a, y-b)

    m5 = (x+b, y+a)
    m6 = (x-b, y+a)
    m7 = (x+b, y-a)
    m8 = (x-b, y-a)
    
    moves = [m1,m2,m3,m4,m5,m6,m7,m8] 
    possible_moves = []
    for item in moves:
        if item[0]<0 or item[1]<0 or item[0]>n-1 or item[1]>n-1:
            None
        else:
            possible_moves.append(item)
    
    return possible_moves
```

```python
def knight(x,y,a,b,n,goal):
    '''
    A function that applied breadth-first search to get to a goal position
    on an nxn grid starting from (x,y) point on the grid with allowed moves
    (a,b) for each step. The output is the possible visited points, the parent
    points for each visited point and the number of steps to reach that point.
    It also prints out a staement when the goal is reached with the number of
    steps needed, which is the shortest path.
    '''
    visited = [(x,y)]
    Q = [(x,y)]
    parent = [None]
    nb_moves = [0]

    while len(Q) != 0 :
        u = Q[0]
        Q.pop(0)
        if u == goal:
            print('Goal Reached with %d steps with a = %d and b = %d'% ((nb_moves[visited.index(goal)]),a,b))
        moves = move(u[0],u[1],a,b,n)
        for node in moves:
            if node not in visited:
                visited.append(node)
                parent = parent + [u]
                nb_moves.append(nb_moves[visited.index((parent[-1][0],parent[-1][1]))]+1)
                Q.append(node)
    return visited, parent, nb_moves
```

```python
x , y = 0, 0
# a , b = 1, 1
n = 25
goal = (n-1,n-1)
itr = 0
sum_moves = 0
for b in range(1,25):
    for a in range(1,b+1):
        visited, parent, nb_moves = knight(x,y,a,b,n,goal)
        
        if goal not in visited:
            print('Never Possible for a= %d and b= %d'%(a,b))
            itr = itr+1
        else:
            sum_moves = sum_moves+nb_moves[visited.index((24,24))]
```
